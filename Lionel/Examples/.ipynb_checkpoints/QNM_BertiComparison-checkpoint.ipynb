```python
# Disable Print
def blockPrint():
    sys.stdout = open(os.devnull, 'w')

# Restore Print
def enablePrint():
    sys.stdout = old_stdout
    
#Correct phase from arctangent
def func_phase(varphase):

        varphi = np.copy(varphase)
        for i in range(len(varphase)):
            if abs(varphase[i-1]-varphase[i]-2.*np.pi)<0.1:
                varphi[i:] = varphi[i:] + 2.*np.pi
        return varphi


#Compute the local maximum of any (normal behaved) function
def local_max(a):
    max_a = []
    max_a_it = []
    for i in range(1,len(a)-1):
        if ((a[i]>a[i+1]) and a[i]>a[i-1]):
            max_a.append(a[i])
            max_a_it.append([i])
    return max_a, max_a_it

#Compute the local minimum of any (normal behaved) function
def local_min(a):
    min_a = []
    min_a_it = []
    for i in range(1,len(a)-1):
        if ((a[i]<a[i+1]) and a[i]<a[i-1]):
            min_a.append(a[i])
            min_a_it.append([i])
    return min_a, min_a_it

    
def func_qnm_amp(time, amp, phi):
    
    #Fitting:
    t_max_amp = time[np.where(amp==np.amax(amp))][0]
    t1_idx = np.amin(np.where(time>=t_max_amp+40))
    t2_idx = np.amin(np.where(time>=t_max_amp+80)) #After t>tmax_amp+90M,  oscillations have been observed in log(amp)
    log_amp = np.log(amp)
    logamp_fit = np.poly1d(np.polyfit(time[t1_idx:t2_idx], log_amp[t1_idx:t2_idx], 1))
        
    phi_fit = np.poly1d( np.polyfit(time[t1_idx:t2_idx], phi[t1_idx:t2_idx], 1))
    
    decay_t = -1./logamp_fit.c[0]
    qnm_freq = phi_fit.c[0]
    
    return  decay_t, qnm_freq, phi_fit, logamp_fit

def func_qnm_real(time, amp, phi):
    
    #Fitting:
    
    t_max_amp = time[np.where(np.abs(amp)==np.amax(np.abs(amp)))][0]
    t1_idx = np.amin(np.where(time>=t_max_amp+40))
    t2_idx = np.amin(np.where(time>=t_max_amp+80)) #After t>tmax_amp+90M,  oscillations have been observed in log(amp)
    log_amp = np.log(amp)
    logamp_fit = np.poly1d(np.polyfit(time[t1_idx:t2_idx], log_amp[t1_idx:t2_idx], 1))
        
    phi_fit = np.poly1d( np.polyfit(time[t1_idx:t2_idx], phi[t1_idx:t2_idx], 1))
    
    decay_t = -1./logamp_fit.c[0]
    qnm_freq = phi_fit.c[0]
    
    return  decay_t, qnm_freq, phi_fit, logamp_fit
```
