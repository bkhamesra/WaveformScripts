# Creating fake signal and checking Reisswig's  method

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate 
get_ipython().magic(u'matplotlib inline')
```

```python
t = np.arange(0,1000,0.1)

y_an = np.sin(2.* np.pi*t/100)
random = np.random.uniform(-1.e-3,1.e-3, len(t))
print random
y_num = y_an + random

plt.figure(figsize=(8,6))
plt.plot(t, y_num, 'r--', lw=2, label="Numerical Solution")
plt.plot(t, y_an, 'b', lw=1, label="Analytic Solution")
plt.ylim(-2,2)
plt.xlabel('Time')
plt.ylabel('f(t)')
plt.grid(True)
plt.legend()
plt.show()
plt.close()
```
