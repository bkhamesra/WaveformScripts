```python
import numpy as np
import matplotlib.pyplot as plt
#get_ipython().magic(u'matplotlib inline')
```

```python
str_22 = '/nethome/numrel/datafiles/Finalized_Waveforms/Waveform_files/Completed/Old/NonSpinning/D11_q1.50_a0.0_m200/Strain_l2_m2.txt'
psi4_22 = '/nethome/numrel/datafiles/Waveforms/HR-series/q1.50/D11_q1.50_a0.0_m200/Ylm_WEYLSCAL4::Psi4r_l2_m2_r75.00.asc'
```

```python
t1, amp_str = np.loadtxt(str_22, unpack=True, usecols=(0,3))
t2, re, im= np.loadtxt(psi4_22, unpack=True, usecols=(0,1,2))
amp_psi4 = np.sqrt(re**2 + im**2) 

tmax_str = t1[amp_str==np.amax(amp_str)]
tmax_psi = t2[amp_psi4==np.amax(amp_psi4)]
print tmax_psi, tmax_str
```

```python
f, (ax1, ax2) = plt.subplots(2,1)
ax1.plot(t1, amp_str, 'b', label='Strain')
ax2.plot(t2, amp_psi4, 'r', label='Psi4')

starty,endy = ax1.get_ylim()
ax1.plot([tmax_str,tmax_str], [starty,np.amax(amp_str)], 'k--', linewidth =1.5)
ax1.text(tmax_str,np.amax(amp_str)+0.00003,'%d'%int(tmax_str), horizontalalignment='center', fontsize=12)
starty,endy = ax2.get_ylim()
ax2.plot([tmax_psi,tmax_psi], [starty,np.amax(amp_psi4)], 'k--', linewidth =1.5)
ax2.text(tmax_psi,np.amax(amp_psi4)+0.00003,'%d'%int(tmax_psi), horizontalalignment='center', fontsize=12)


ax2.set_xlabel('Time')
ax1.set_ylabel('Strain')
ax2.set_ylabel('Psi4')
#plt.legend()
plt.show()
plt.close()
```
